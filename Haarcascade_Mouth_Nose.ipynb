```python
def face_Detect(image):
  #image=cv2.resize(image,(800,900))
  nose_detector=cv2.CascadeClassifier("/content/drive/MyDrive/Thesis_detection/CascaseFiles/haarcascade_mcs_nose.xml")
  mouth_detector=cv2.CascadeClassifier("/content/drive/MyDrive/Thesis_detection/CascaseFiles/haarcascade_mcs_mouth.xml")
  face_detector=cv2.CascadeClassifier("/content/drive/MyDrive/Thesis_detection/CascaseFiles/haarcascade_frontalface_default.xml")
  #detection1 = face_detector.detectMultiScale(image,scaleFactor=1.08,minNeighbors=6)
  face = face_detector.detectMultiScale(image,1.02,minSize=(150,150))
  mouth = mouth_detector.detectMultiScale(image, 1.3)
  nose = nose_detector.detectMultiScale(image, 1.01,minNeighbors=5)
  print(face)
  print(mouth)
  print(nose)
  for (x,y,w,h) in face:
    #print(x,y,w,h)
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2)

  for (mx,my,mw,mh) in mouth:
    #print(mx+mw,my+mh)
    cv2.rectangle(image,(mx,my),(mx+mw,my+mh),(0,255,0),2)

  for (nx,ny,nw,nh) in nose:
    cv2.rectangle(image,(nx,ny),(nx+nh,ny+nw),(255,0,0),2)

  cv2_imshow(image)
  '''
  #ndetect=nose_detector.detectMultiScale(image)
  mdetect=mouth_detector.detectMultiScale(image)
  for (x,y,w,h) in ndetect:
    cv2.rectangle(image,(x,y),(x+h),(y+w),(0,255,0),2)
  for (x,y,w,h) in mdetect:
    cv2.rectangle(image,(x,y),(x+h),(y+w),(0,0,255),2)
from google.colab.patches import cv2_imshow
'''
#cv2_imshow(image)
```
