```python
import glob
import cv2

images = [cv2.imread(file) for file in glob.glob("/content/drive/MyDrive/Thesis_detection/Images/*.png")]
from google.colab.patches import cv2_imshow
p="/content/drive/MyDrive/Thesis_detection/dlib/shape_predictor_68_face_landmarks.dat"
```

```python
def face_detect_dlib(image):
  detector = dlib.get_frontal_face_detector()
  predictor= dlib.shape_predictor(p)
  faces = detector(image)
  for face in faces:
    l,t,r,b=face.left(),face.top(),face.right(),face.bottom()
    cv2.rectangle(image,(l,t),(r,b),(255,255,0),2)
    landmarks= predictor(image,face)
    for n in range(0,68):
      x=landmarks.part(n).x
      y=landmarks.part(n).y
      cv2.circle(image, (x, y), 4, (0, 0, 255), -1)
  cv2_imshow(image)
```

```python
for image in images:
  face_detect_dlib(image)
```
